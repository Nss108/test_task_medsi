## №1 
По данным из файла "Загрузка_июнь 2022 test" посчитайте месячную загрузку Отделения по должностям, как процент времени приёма от времени на рабочем месте

```python
df_zagruzka['YearMonth'] = df_zagruzka['VisitDate'].dt.to_period('M')


grouped_data = df_zagruzka.groupby(['Актив','Отделение', 'Специальность', 'YearMonth']).agg(
    total_work_time=('Длительность с учетом перерыва (мин.)', 'sum'),
    total_appointment_time=('Фактическая загрузка (в мин.)', 'sum')
).reset_index()

grouped_data['load_percentage'] = (grouped_data['total_appointment_time'] / grouped_data['total_work_time'] * 100).round(2)

grouped_data.head(20)
```

## №2 
Объедините данные данные из файла "Штатная расстановка test" и результат выполнения задания №1

```python
df_sh = df_sh.rename(columns={'Должность': 'Специальность'})
df_sh = pd.merge(df_sh, grouped_data, on=['Актив', 'Отделение', 'Специальность'], how='inner')
df_sh.head(80)
```

## №3 
Добавьте столбец в объединенную таблицу из задания №2 по следующему условию: если загрузка отделения выше или равна 60%, то "+", если ниже, то "-"

```python
df_sh['LoadIndicator'] = df_sh['load_percentage'].apply(lambda x: '+' if x >= 60 else '-')
df_sh.head(5)
```

## №4
Проанализируйте, сколько ставок (кратно 0,25) можно добавить в отделения с загрузкой выше или равной 60%, чтобы сделать её ниже 60%. В новый столбец запишите, сколько ставок можно добавить. (Месячная норма на 1 ставку для всех должностей счиать равной 160 часов)

```python
one_job_minutes = 160 * 60
df_sh['JobsTo60'] = np.round(((df_sh['total_appointment_time'] / 0.6) - df_sh['total_work_time']) / one_job_minutes * 4) / 4

df_sh.tail(40)
```

## №5
В новый столбец пропишите текстовый комментарий, включающий в себя следующие части:
Оклад на 1 ставку, Административная премия
Целевой бонус
Премия за результат
% Касса
% Факт (СК)
Премия за авансовых пациентов
Целевая премия на 1 ставку
Целевой доход на 1 ставку

```python
df_sh.columns = df_sh.columns.str.replace("\n", " ")
df_sh.columns = df_sh.columns.str.strip()
def generate_comment(row):
    comment = f"Оклад на 1 ставку: {row['Оклад на 1 ставку, руб.']}, "
    comment += f"Административная премия: {row['Административная премия, %']}, "
    comment += f"Целевой бонус: {row['Целевой бонус, %']}, "
    comment += f"Премия за результат: {row['Премия за результат %']}, "
    comment += f"Касса: {row['% Касса - …. %']}, "
    comment += f"Факт (СК): {row['% Факт (СК) - …. %']}, "
    comment += f"Премия за авансовых пациентов: {row['Премия за авансовых пациентов']}, "
    comment += "Целевая премия на 1 ставку: N/A, "
    comment += "Целевой доход на 1 ставку: N/A"
    
    return comment

df_sh['Comment'] = df_sh.apply(generate_comment, axis=1)

df_sh.head()
```
